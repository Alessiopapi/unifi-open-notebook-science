<h2> funzione per il calcolo dell'indice di gini </h2>

```python
def gini(series):
    """Calculate the Gini coefficient of a numpy array."""
    # based on bottom eq: http://www.statsdirect.com/help/content/image/stat0206_wmf.gif
    # from: http://www.statsdirect.com/help/default.htm#nonparametric_methods/gini.htm
    # from Oliviaguest repository: https://github.com/oliviaguest/gini
    series = series.apply(lambda x: float(x))
    array =  series.values
    array = array.flatten() #all values are treated equally, arrays must be 1d
    if np.amin(array) < 0:
        array -= np.amin(array) #values cannot be negative
    array += 0.0000001 #values cannot be 0
    array = np.sort(array) #values must be sorted
    index = np.arange(1,array.shape[0]+1) #index per array element
    n = array.shape[0]#number of array elements
    return ((np.sum((2 * index - n  - 1) * array)) / (n * np.sum(array))) #Gini coefficient
```

```python
gini(Italia)
```

```python
gini(Svezia)
```

```python
gini(Usa)
```

```python
#riprendo la tabella pulita "dati grezzi"
tabella = pd.DataFrame([Italia, Svezia, Usa])
tabella.columns =  columns = ["I-Q", "II-Q", "III-Q", "IV-Q", "V-Q"]
tabella
```

```python
gini_series = tabella.apply(gini, axis = 1) #applico la funzione gini a tutte le righe
gini_series.name = "gini"
gini_series
```
