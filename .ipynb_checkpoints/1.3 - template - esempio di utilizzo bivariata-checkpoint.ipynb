```python
# ---------------------------------------------------------------
# python importa la libreria OS che controlla il sistema operativo
import os 
# ---------------------------------------------------------------
# python importa la libreria Pandas che permette l'importazione ed analisi dei dati: potrà essere richiamata come PD
import pandas as pd 

#importo le librerie di visualizzazione dei grafici
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#importa la libreria per la pulitura del dataset
from corso_lib import restituisci_dataset
```

```python
'''
PRIMO DATASET
Tavola 3.1 segue - Persone di 30-34 anni che hanno conseguito 
un titolo universitario per sesso, regione e ripartizione geografica - Anni 2004-2015 (valori percentuali)
'''

totale_istruzione = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "02.Istruzione", "Cap_02_Istruzione 03.xlsx"),
    nome_del_foglio = "REG totale", #inserire il nome del foglio
    riga_intestazione_colonna = 4, # inserire la riga in cui è presente l'intestazione di colonna 
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
totale_istruzione
```

```python
'''
SECONDO DATASAET
REG totale: Tavola 5.1 segue - Incidenza di lavoratori dipendenti con bassa paga per sesso, regione e ripartizione geografica per sesso, 
regione e ripartizione geografica - Anni 2008-2015 (valori percentuali)
os.path.join(os.getcwd(), "Appendice-statistica-2016", "03.Lavoro", "Cap_03_Lavoro 05.xlsx")
"REG totale"
'''
```

```python
lavoratori_bassa_paga = restituisci_dataset(
    posizione_del_dataset = os.path.join(os.getcwd(), "Appendice-statistica-2016", "03.Lavoro", "Cap_03_Lavoro 05.xlsx"),
    nome_del_foglio = "REG totale",
    riga_intestazione_colonna = 4,
    riga_sottotabella= None, # None - da divisione fra le due tabelle "maschi - femmine" dovrebbe cadere intorno alla riga 36
    tabella = None # None - qui si può mettere "maschi" - "femmine"
)
```

```python
#seleziono e concateno gli anni più recenti
istruzione_su_bassa_paga = pd.concat([totale_istruzione[2015], lavoratori_bassa_paga[2015]], axis = 1)
istruzione_su_bassa_paga
```

```python
#aggiungo i nomi delle colonne
istruzione_su_bassa_paga.columns = ["30-34 anni che hanno conseguito un titolo universitario", "Incidenza di lavoratori dipendenti con bassa paga - 2015"]
istruzione_su_bassa_paga
```

```python
#Analisi bivariata - correlazione fra le due variabili
istruzione_su_bassa_paga.corr()
```

```python
#visaulizzazione base (scatterplot) di pandas + annotazioni

fig, ax = plt.subplots()

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = istruzione_su_bassa_paga
asse_x = "30-34 anni che hanno conseguito un titolo universitario"
asse_y = "Incidenza di lavoratori dipendenti con bassa paga - 2015"
### ----------------------------------------------------------------

dataset.plot(asse_x, asse_y, kind='scatter', ax=ax, figsize = [15,10])

for k, v in dataset.iterrows():
    ax.annotate(k, v)
    
fig.canvas.draw()
```

```python
#grafico di seaborn (scatterplot) di pandas + annotazioni
#se non fosse installata la libreria seaborn
# da terminale:    pip install seaborn 

import seaborn as sns


sns.set(style="darkgrid")

### VARIABILI DA PERSONALIZZARE CON I PROPRI DATI
dataset = istruzione_su_bassa_paga
asse_x = "30-34 anni che hanno conseguito un titolo universitario"
asse_y = "Incidenza di lavoratori dipendenti con bassa paga - 2015"
nome_grafico = "uni__bassapaga.png"
### ----------------------------------------------------------------


color = sns.color_palette()[2]
sns.jointplot(asse_x, asse_y, data=dataset, kind="reg",
                   color=color, size=7) #xlim=(0, 60), ylim=(0, 12)

def ann(row):
    ind = row[0]
    r = row[1]
    plt.gca().annotate(ind, xy=(r[asse_x], r[asse_y]), 
            xytext=(2,2) , textcoords ="offset points", )

for row in dataset.iterrows():
    ann(row)    

fig = matplotlib.pyplot.gcf()
fig.set_size_inches(20, 15)
plt.savefig("uni__bassapaga.png", dpi=300)
```
