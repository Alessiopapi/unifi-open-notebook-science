```python
#importiamo la liberia che permette di navigare sulle cartelle del PC
import os

#importo la libreria statistica pandas
import pandas as pd

#importa la libreria per manipolare fie HTML (formato delle pagine del WEB)
from bs4 import BeautifulSoup
```

```python
#carichiamo il programma in memoria per selezionare il dominio ed estrarre i dataset

with open("indice dei dataset bes.html", "w") as indice_html:
            indice_html.write("<h1>" + "Lista dei dataset presenti" + "</H1>")    



def crezione_indice():
    #print(os.getcwd())
    lista_domini = os.listdir(os.path.join(os.getcwd(), 'Appendice-statistica-2016'))
    dizionario_per_selezione = {}
    for dominio in lista_domini:
        #print(dominio)
        with open("indice dei dataset bes.html", "a") as indice_html:
            indice_html.write("<h1>" + dominio + "</h1>")

        for excel_name in os.listdir(os.path.join(os.getcwd(), 'Appendice-statistica-2016', dominio)):
            #print(excel_name)
            with open("indice dei dataset bes.html", "a") as indice_html:
                indice_html.write("<h2>" + excel_name + "</h2>")
            sheets= pd.read_excel(os.path.join(os.getcwd(), 'Appendice-statistica-2016', dominio, excel_name), sheetname = None)
            link = 'os.path.join(os.getcwd(), "Appendice-statistica-2016", "{}", "{}")'.format(dominio, excel_name)
            #print(os.getcwd(), 'Appendice-statistica-2016', dominio, excel_name)
            with open("indice dei dataset bes.html", "a") as indice_html:
                indice_html.write("<h3>" + str(link) + "</h3>")
            for sheet in sheets.keys():
                try:
                    #print(str(sheets[sheet].iloc[1,0]) + " " + str(sheets[sheet].iloc[2,0]))
                    with open("indice dei dataset bes.html", "a") as indice_html:
                        indice_html.write("<p>"+ str(sheet) + ": " + str(sheets[sheet].iloc[1,0]) + " " + str(sheets[sheet].iloc[2,0]) + "</p>")
                    
                    
                    #print(sheet)
                except:
                    print(sheet + "--- error --------------------------------------------------------------------------")
```

```python
crezione_indice()
```
